# Product Identification

```python
def text_meta_filter(str_contain, str_not, items_meta):
    """
    takes a data set and string filters and returns a filtered dataset
    """
    for string in str_contain:
        items_meta = items_meta[items_meta['item_name'].str.contains(string)] # keep only items that contain tomato string
    for string in str_not:
        items_meta = items_meta[~items_meta['item_name'].str.contains(string)] # keep only items that contain tomato string
    return items_meta
```

```python
def get_prod_ids(filtered_items, chains_id):
    """
    takes a dataset and a chain list and returns a list of the top item occurences in each chain 
    """
    product_id_table = pd.DataFrame()
    for c_id in chains_id:

        print("current chain id: %s" %c_id)

        # get prices updates of product items from current chain stores
        cur_items_id = filtered_items.loc[filtered_items['chain_id'] == c_id].loc[filtered_items['is_weighted'] == '1']['item_id'] # get all weighted item_id of current chain
        prices = pd.read_sql_query("SELECT * from prices where item_id in %s" %str(tuple(cur_items_id)) , db)

        current_items = prices['item_id'].value_counts().to_frame().assign(chain_id=c_id) # count timestamp updates
        if len(current_items) != 0:
            product_id_table = product_id_table.append(current_items.iloc[[0]]) # keep only top occurence
        
    product_id_table['item_count'] = product_id_table['item_id']
    product_id_table['item_id'] = product_id_table.index
    
    return product_id_table
```

```python
# explore product in every store and get item_id_table of product
str_contain = ['עגבני']
str_not = ['שרי', 'אורגני','רסק', 'ליקופן']
items_meta = pd.read_sql_query("SELECT * from items_meta", db)

filtered_items = text_meta_filter(str_contain, str_not, items_meta)

prod_id_table = get_prod_ids(filtered_items, chains['chain_id'])

# merge details from other table in the db
prod_id_table = prod_id_table.merge(chains, on='chain_id', how='left')
prod_id_table = prod_id_table.merge(items_meta[['chain_id','item_id','item_name']], on= ['chain_id', 'item_id'], how = 'left').drop_duplicates(subset='item_id')
```

# build a panel from list of items

```python
# create panel data
def create_panel(items_list):
    '''
    takes a list of valid items and returns a panel of the items prices in each store in th DB
    * panel does not contain stores that didnt update the item price for more than 100 days
    ** panel does not contain stores that didnt change their price in 2016
    '''
    top = datetime.datetime.strptime('01012016', "%d%m%Y").date()
    
    all_dates = pd.DataFrame({'timestamp': pd.date_range(start='20/5/2015', end='31/12/2016')})
    all_dates['timestamp'] = all_dates['timestamp'].dt.date

    all_prices = pd.read_sql_query("SELECT * from prices where item_id in %s" % str(tuple(items_list)), db)
    all_prices['timestamp'] = pd.to_datetime(all_prices['timestamp'], unit='s').dt.date
    
    all_chains = pd.DataFrame()
    bad_stores = pd.DataFrame()
    t = datetime.timedelta(days=100)
    for item_id in items_list:
        print(item_id)
        curr_chain = all_prices.loc[all_prices['item_id'] == item_id]
        chain_df = pd.DataFrame()

        # create the data for each store in current chain
        for store_id in curr_chain['store_id'].sort_values().unique():
            curr_store = curr_chain.loc[curr_chain['store_id'] == store_id]
            curr_diff = curr_store['timestamp'].diff().max()
            
            # do not merge stores that didnt update their prices for more than 100 days
            if type(curr_diff) != float:
                if curr_diff.days > 100:
                    print(store_id, ": ", curr_store['timestamp'].diff().max())
                    continue
            
            curr_store = all_dates.merge(curr_store, on='timestamp', how='left')
            curr_store = curr_store.fillna(method='ffill') # fill in price for dates that the price wasnt updated
            
            # do not add stores that have the same value 
            if curr_store.loc[curr_store['timestamp'] > top]['price'].value_counts().max() > 200:
                print(store_id, " to many values")
                bad_stores = bad_stores.append(curr_store).dropna()
                bad_stores.index = bad_stores.timestamp
           
            else:
                chain_df = chain_df.append(curr_store).dropna()
                chain_df.index = chain_df.timestamp
        
        chain_df.to_csv('panel_%s.csv' %item_id , index_label = 'index')
        all_chains = all_chains.append(chain_df)
    
    all_chains = all_chains[['timestamp', 'item_id', 'store_id', 'price']]
    return all_chains, bad_stores
```

# Data Science

I will try to predict the type of the store according to the price of tomatoes

```python
mega = my_df[my_df['chain_name'] == 'Mega']
```

```python
mega
```

```python
# create a data s.t every store is a row, each price at a timestamp is a feature and the type of the stores are the labels
# only mega has valid store types data about its stores. (mega ba'air and Mega zul begadol)
df = pd.DataFrame()
for store_id in mega['store_id'].drop_duplicates():
    df[str(store_id)] = mega[mega['store_id'] == store_id]['price']
labels = mega[['store_id', 'subchain_id']].drop_duplicates().sort_values(by='store_id')

df.index.names = ['store_id']
df = df.transpose()
df['store_id'] = df.index.astype(float)
data = df.merge(labels, on='store_id', how='left').drop('store_id' , axis=1)
data = data.dropna()

pd.set_option('display.max_rows', 30)
data
```

## A simple KNeighborsClassifier will do the trick (it's a simple task)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1],  data.iloc[:,-1:], test_size=0.40, random_state=42)
```

```python
neigh = KNeighborsClassifier(n_neighbors=10)
neigh.fit(X_train, y_train)
```

```python
neigh.predict(X_test)
```

```python
neigh.score(X_test, y_test.iloc[:,0], sample_weight=None)
```

```python
pd.set_option('display.max_rows', 80)
print(pd.DataFrame({'prediction': neigh.predict(X_test),'true_labels': y_test.iloc[:,0]}))
```
